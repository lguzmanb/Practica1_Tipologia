```python
import requests
from bs4 import BeautifulSoup
import csv

url= "https://www.ikea.com/es/es/productos/camas/cabeceros-cama/"

#Descarregar pagina web
page=requests.get(url)

#Contingut en brut de la pàgina
soup=BeautifulSoup(page.content)

#Estructura per detectar la informació d'interès
soup_pretty=soup.prettify()

#Mirem estructura
print ("Pàgina desglosada/n",soup_pretty)


```
